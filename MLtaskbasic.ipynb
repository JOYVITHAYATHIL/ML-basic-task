```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/dataset/train_updated.csv")
df2 = pd.read_csv("/content/drive/MyDrive/dataset/test.csv")
```

```python
arr = df1.to_numpy()
tset = df2.to_numpy()
m = arr.shape[0]
w = 0.5
b = 0.5
x = arr[:,0]
y = arr[:,1]
mt = tset.shape[0]
x1 = tset[:int(0.5*mt),0]
y1 = tset[:int(0.5*mt),1]
x2 = tset[int(0.5*mt):,0]
y2 = tset[int(0.5*mt):,1]
m1 = x1.shape[0]
m2 = x2.shape[0]
l = []
for iter in range(1000):
  yh = w*x+b
  e = yh- y
  w -= 0.01/m*(sum(w*e))
  b -= 0.01/m*(sum(e))
  print('iter = ',iter+1)
  print('train loss = ',(1/(2*m))*np.sum(e**2))
  yh = w*x1+b
  e = yh- y1
  l += [(1/(2*m1))*np.sum(e**2)]
  print('cross validation loss = ',(1/(2*m1))*np.sum(e**2))
  print('----------------------------------------------------')
yh = w*x2+b
e = yh- y2
j = (1/(2*m2))*np.sum(e**2)
print('\n\ntest loss = ',j)
plt.plot(np.arange(1,1001),l)
plt.title("Learning curve")
plt.ylabel("Loss")
plt.xlabel("Number of iterations")
plt.show()
plt.scatter(x2,y2)
plt.ylabel("x")
plt.xlabel("y")
plt.title("Actual vs Prediction")
plt.plot(x2,yh,color = "red")
plt.show()
```

```python
print("R2 score :",1 - sum(e**2)/sum((y2-np.mean(y2))**2))
print("Root mean square loss of test dataset : ",(2*j)**0.5)
```
